```python
# import potřebných knihoven

import pandas as pd
import re
import numpy as np
import csv
from pathlib import Path
```

```python
# definice cest k datovým souborům

DATA_DIR = Path(".")

VSE_FILE = DATA_DIR / "vse_syllabi.csv"

PARTNER_FILES = [
    DATA_DIR / "queens_syllabi.csv",
    DATA_DIR / "canterbury_syllabi.csv",
    DATA_DIR / "ITAM_syllabi.csv",
    DATA_DIR / "hkust_syllabi.csv",
    DATA_DIR / "edinburgh_syllabi.csv",
    DATA_DIR / "bi_syllabi.csv"
]
```

```python
# načtení datové sady sylabů VŠE

df_vse = pd.read_csv(
    VSE_FILE,
    dtype=str
).fillna("")

df_vse.shape
```

```python
# načtení a spojení datových sad partnerských univerzit

partner_dfs = []

for file in PARTNER_FILES:

    df = pd.read_csv(
        file,
        dtype=str
    ).fillna("")


    partner_dfs.append(df)

df_partners = pd.concat(
    partner_dfs,
    ignore_index=True
)

df_partners.shape
```

```python
# kontrola názvů sloupců v obou datových sadách

print("VŠE sloupce:")
print(df_vse.columns.tolist())

print("\nPartnerské univerzity sloupce:")
print(df_partners.columns.tolist())
```

```python
# základní kontrola velikosti datových sad a zastoupení univerzit

print("VŠE:", df_vse.shape)

print("Partnerské univerzity:", df_partners.shape)

df_partners["university"].value_counts()
```

```python
# základní kontrola prázdných hodnot ve všech sloupcích

def count_empty_values(df):
    return (
        df.eq("")
        .sum()
        .sort_values(ascending=False)
    )

print("VŠE:")
display(count_empty_values(df_vse))

print("Partnerské univerzity:")
display(count_empty_values(df_partners))
```

```python
# kontrola atributu course_title

COLUMN = "course_title"

def check_titles(df, dataset_name):

    print(f"\n{'='*60}")
    print(f"{dataset_name}")
    print(f"{'='*60}")

    print("\nPočet prázdných hodnot:")
    print((df[COLUMN] == "").sum())

    print("\nNejkratší názvy:")
    display(
        df[[COLUMN]]
        .assign(length=df[COLUMN].str.len())
        .sort_values("length")
        .head(10)
    )

    print("\nNáhodný vzorek:")
    display(
        df[COLUMN]
        .sample(min(10, len(df)), random_state=42)
    )

check_titles(df_vse, "VŠE")
check_titles(df_partners, "Partnerské univerzity")
```

2. course_code

```python
# kontrola atributu course_code

COLUMN = "course_code"

def check_course_codes(df, dataset_name):

    print(f"\n{'='*60}")
    print(f"{dataset_name}")
    print(f"{'='*60}")

    print("\nPočet prázdných hodnot:")
    print((df[COLUMN] == "").sum())

    print("\nPočet unikátních kódů:")
    print(df[COLUMN].nunique())

    print("\nDuplicitní kódy:")
    duplicates = (
        df[df[COLUMN].duplicated(keep=False)]
        .sort_values(COLUMN)
    )

    if duplicates.empty:
        print("Žádné duplicity.")
    else:
        display(
            duplicates[
                ["course_code", "course_title", "university"]
            ]
        )

    print("\nNáhodný vzorek:")
    display(
        df[COLUMN]
        .sample(min(10, len(df)), random_state=42)
    )

check_course_codes(df_vse, "VŠE")

check_course_codes(df_partners, "Partnerské univerzity")
```

```python
# kontrola, zda se u duplicitních course_codes jedná o celé duplicitní řádky

duplicate_codes = (
    df_partners.loc[df_partners["course_code"].duplicated(keep=False), "course_code"]
    .unique()
)

duplicate_check = []

for code in duplicate_codes:
    rows = df_partners[df_partners["course_code"] == code]
    
    unique_full_rows = rows.drop_duplicates()
    
    duplicate_check.append({
        "course_code": code,
        "row_count": len(rows),
        "unique_full_rows": len(unique_full_rows),
        "same_full_row": len(unique_full_rows) == 1,
        "university": rows["university"].iloc[0],
        "course_title": rows["course_title"].iloc[0]
    })

duplicate_check_df = pd.DataFrame(duplicate_check)

display(
    duplicate_check_df.sort_values(
        ["same_full_row", "course_code"]
    )
)
```

```python
# kontrola dvou řádků se stejným kódem
df_partners[
    df_partners["course_code"] == "COMM 200"
].T
```

```python
# kontrola duplicit course_code mimo semester atributy

COMPARE_COLUMNS = [
    col for col in df_partners.columns
    if col not in ["semester", "semester_normalized"]
]

duplicate_codes = (
    df_partners.loc[
        df_partners["course_code"].duplicated(keep=False),
        "course_code"
    ]
    .unique()
)

duplicate_check = []

for code in duplicate_codes:

    rows = df_partners[
        df_partners["course_code"] == code
    ]

    unique_rows = (
        rows[COMPARE_COLUMNS]
        .drop_duplicates()
    )

    duplicate_check.append({
        "course_code": code,
        "row_count": len(rows),
        "unique_rows_without_semester": len(unique_rows),
        "same_except_semester": len(unique_rows) == 1,
        "university": rows["university"].iloc[0],
        "course_title": rows["course_title"].iloc[0]
    })

duplicate_check_df = pd.DataFrame(duplicate_check)

display(
    duplicate_check_df.sort_values(
        ["same_except_semester", "course_code"]
    )
)
```

```python
# detailní kontrola duplicitního kódu, který se liší i mimo semester

df_partners[
    df_partners["course_code"] == "COMM 321"
].T
```

```python
# nalezení duplicitních course_code u Queen's University

duplicate_codes_queens = (
    duplicate_check_df.loc[
        duplicate_check_df["university"] == "Queen's University",
        "course_code"
    ]
    .tolist()
)

# odstranění Fall variant u obsahově stejných kurzů

mask_queens_fall_duplicates = (
    (df_partners["university"] == "Queen's University") &
    (df_partners["course_code"].isin(duplicate_codes_queens)) &
    (df_partners["semester"] == "Fall")
)

df_partners = (
    df_partners
    .loc[~mask_queens_fall_duplicates]
    .reset_index(drop=True)
)

# odstranění plně identických duplicit

df_partners = (
    df_partners
    .drop_duplicates()
    .reset_index(drop=True)
)

print(df_partners.shape)
```

```python
# kontrola atributu source_url

COLUMN = "source_url"

def check_source_url(df, dataset_name):

    print(f"\n{'='*60}")
    print(f"{dataset_name}")
    print(f"{'='*60}")

    print("\nPočet prázdných hodnot:")
    print((df[COLUMN] == "").sum())

    print("\nPočet unikátních URL:")
    print(df[COLUMN].nunique())

    print("\nPočet duplicitních URL:")
    duplicate_count = (
        df[COLUMN]
        .duplicated(keep=False)
        .sum()
    )

    print(duplicate_count)

    if dataset_name == "Partnerské univerzity":

        print("\nDuplicitní URL podle univerzit:")

        duplicates_by_university = (
            df.loc[
                df[COLUMN].duplicated(keep=False)
            ]
            .groupby("university")
            .size()
            .sort_values(ascending=False)
        )

        display(duplicates_by_university)

    print("\nNáhodný vzorek URL:")
    display(
        df[COLUMN]
        .sample(min(10, len(df)), random_state=42)
    )

check_source_url(df_vse, "VŠE")

check_source_url(df_partners, "Partnerské univerzity")
```

```python
# kontrola prefixů course_code u HKUST

hkust_codes = df_partners[
    df_partners["university"] == "The Hong Kong University of Science and Technology"
].copy()

hkust_codes["code_prefix"] = (
    hkust_codes["course_code"]
    .astype(str)
    .str.strip()
    .str.extract(r"^([A-Z]+)")
)

print("Počet prázdných prefixů:")
print(hkust_codes["code_prefix"].isna().sum())

print("\nPočty předmětů podle prefixu:")
display(
    hkust_codes["code_prefix"]
    .value_counts()
    .sort_index()
)
```

```python
# zpřesnění URL u HKUST podle prefixu course_code

def update_hkust_url(row):
    if row["university"] != "The Hong Kong University of Science and Technology":
        return row["source_url"]

    code = str(row["course_code"]).strip()

    match = re.match(r"^([A-Z]+)", code)

    if match:
        subject_prefix = match.group(1)
        return f"https://prog-crs.hkust.edu.hk/ugcourse/2025-26/{subject_prefix}"

    return row["source_url"]


df_partners["source_url"] = df_partners.apply(
    update_hkust_url,
    axis=1
)
```

6. academic_year

```python
# kontrola atributu academic_year

COLUMN = "academic_year"

def check_academic_year(df, dataset_name):

    print(f"\n{'='*60}")
    print(f"{dataset_name}")
    print(f"{'='*60}")

    print("\nPočet prázdných hodnot:")
    print((df[COLUMN] == "").sum())

    print("\nPočet unikátních hodnot:")
    print(df[COLUMN].nunique())

    print("\nUnikátní hodnoty:")
    display(
        df[COLUMN]
        .value_counts(dropna=False)
        .sort_index()
    )

check_academic_year(df_vse, "VŠE")

check_academic_year(df_partners, "Partnerské univerzity")
```

```python
# normalizace academic_year podle semester_normalized

def normalize_academic_year(row):
    year = str(row["academic_year"]).strip()
    semester_normalized = str(row["semester_normalized"]).strip().lower()

    # 2025-2026 nebo 2025/2026 -> 2025/2026
    match_full = re.match(r"^(\d{4})[-/](\d{4})$", year)
    if match_full:
        return f"{match_full.group(1)}/{match_full.group(2)}"

    # 2025-26 nebo 2025/26 -> 2025/2026
    match_short = re.match(r"^(\d{4})[-/](\d{2})$", year)
    if match_short:
        start = int(match_short.group(1))
        end_short = int(match_short.group(2))
        end = int(str(start)[:2] + f"{end_short:02d}")
        return f"{start}/{end}"

    # pokud je uveden jen jeden rok, převod provedeme jen při známém semester_normalized
    match_single = re.match(r"^(\d{4})$", year)
    if match_single:
        y = int(match_single.group(1))

        if semester_normalized == "winter":
            return f"{y}/{y + 1}"

        if semester_normalized == "summer":
            return f"{y - 1}/{y}"

        return year

    return year
```

```python
df_partners["academic_year"] = df_partners.apply(normalize_academic_year, axis=1)
```

```python
# kontrola atributu academic_year

COLUMN = "academic_year"

def check_academic_year(df, dataset_name):

    print(f"\n{'='*60}")
    print(f"{dataset_name}")
    print(f"{'='*60}")

    print("\nPočet prázdných hodnot:")
    print((df[COLUMN] == "").sum())

    print("\nPočet unikátních hodnot:")
    print(df[COLUMN].nunique())

    print("\nUnikátní hodnoty:")
    display(
        df[COLUMN]
        .value_counts(dropna=False)
        .sort_index()
    )

check_academic_year(df_vse, "VŠE")

check_academic_year(df_partners, "Partnerské univerzity")
```

```python
# sjednocení atributu study_level

def normalize_study_level(value):

    value = str(value).strip().lower()

    if value == "master":
        return "master"

    if value == "bachelor":
        return "bachelor"

    if value == "undergraduate":
        return "undergraduate"

    return value


df_vse["study_level"] = (
    df_vse["study_level"]
    .apply(normalize_study_level)
)

df_partners["study_level"] = (
    df_partners["study_level"]
    .apply(normalize_study_level)
)
```

```python
# sjednocení formátu atributu credits

def normalize_credits(value):

    value = str(value).strip()

    if value == "":
        return ""

    try:
        number = float(value)

        # odstranění zbytečné desetinné části
        if number.is_integer():
            return str(int(number))

        return str(number)

    except:
        return value


df_vse["credits"] = (
    df_vse["credits"]
    .apply(normalize_credits)
)

df_partners["credits"] = (
    df_partners["credits"]
    .apply(normalize_credits)
)
```

```python
# kontrola atributu credits po normalizaci

print("VŠE")
display(
    df_vse["credits"]
    .value_counts()
    .sort_index()
)

print("\nPartnerské univerzity")
display(
    df_partners["credits"]
    .value_counts()
    .sort_index()
)
```

```python
# kontrola správnosti přepočtu na ECTS kredity

df_credit_check = df_partners.copy()

df_credit_check["credits_num"] = pd.to_numeric(
    df_credit_check["credits"],
    errors="coerce"
)

df_credit_check["ects_num"] = pd.to_numeric(
    df_credit_check["ects_credits"],
    errors="coerce"
)


def check_ects_conversion(row):

    university = row["university"]
    credits = row["credits_num"]
    ects = row["ects_num"]

    if pd.isna(credits) or pd.isna(ects):
        return False

    if university == "BI Norwegian Business School":
        expected = credits

    elif university == "The Hong Kong University of Science and Technology":
        expected = credits * 2

    elif university == "Instituto Tecnológico Autónomo de México":
        expected = credits

    elif university == "University of Canterbury":
        expected = credits / 2

    elif university == "Queen's University":
        expected = credits * 2

    elif university == "University of Edinburgh":
        expected = credits / 2

    else:
        return False

    return abs(expected - ects) < 0.01


df_credit_check["ects_conversion_correct"] = (
    df_credit_check
    .apply(check_ects_conversion, axis=1)
)

print("Počet správných přepočtů:")
print(
    df_credit_check["ects_conversion_correct"]
    .sum()
)

print("\nPočet nesprávných přepočtů:")
print(
    (~df_credit_check["ects_conversion_correct"])
    .sum()
)

print("\nNesprávné přepočty:")
display(
    df_credit_check[
        df_credit_check["ects_conversion_correct"] == False
    ][
        [
            "course_code",
            "course_title",
            "university",
            "credits",
            "ects_credits",
            "source_url"
        ]
    ]
)
```

```python
# základní čištění atributu learning_outcomes


PLACEHOLDER_VALUES = [
    "The learning outcomes of the course has not been saved in this language version.",
    "-- item not defined --"
]


def clean_learning_outcomes(text):

    # zachování chybějících hodnot
    if pd.isna(text):
        return pd.NA

    # převod na string
    text = str(text)

    # odstranění HTML entit
    text = text.replace("&nbsp;", " ")
    text = text.replace("&amp;", "&")

    # sjednocení odrážek
    text = re.sub(r"[•◦▪■]", "-", text)

    # odstranění nadbytečných mezer a zalomení řádků před kontrolou prázdných hodnot
    text = re.sub(r"\s+", " ", text).strip()

    # placeholdery a prázdné hodnoty -> NaN
    if text in PLACEHOLDER_VALUES:
        return pd.NA

    if text in ["", "-", ".", "_"]:
        return pd.NA

    # odstranění opakujících se úvodních formulací
    boilerplate_patterns = [
        r"^upon successful completion of this course[,]?\s*students will be able to[:]?\s*",
        r"^upon successful completion of the course[,]?\s*students will be able to[:]?\s*",
        r"^upon successful completion[,]?\s*students will be able to[:]?\s*",
        r"^upon completion of this course[,]?\s*students will be able to[:]?\s*",
        r"^upon completion of the course[,]?\s*students will be able to[:]?\s*",
        r"^after completing the course[,]?\s*students will be able to[:]?\s*",
        r"^by the end of the course[,]?\s*students will be able to[:]?\s*",
        r"^by the end of this course[,]?\s*students will be able to[:]?\s*",
        r"^the objectives of the course are[:]?\s*",
        r"^the outcomes of the course are[:]?\s*",
    ]

    for pattern in boilerplate_patterns:
        text = re.sub(pattern, "", text, flags=re.IGNORECASE)

    # odstranění zbývajících boilerplate formulací
    additional_patterns = [
        r"^by the end of the course[,]?\s*",
        r"^by the end of this course[,]?\s*",
        r"^upon completion of the course[,]?\s*students[:]?\s*",
        r"^upon completion of this course[,]?\s*students[:]?\s*",
        r"^after completing the course[,]?\s*the student[:]?\s*",
        r"^after completion of this course[,]?\s*students[:]?\s*",
        r"^upon successful completion of this course[,]?\s*",
        r"^upon successful completion of the course[,]?\s*",
        r"^upon the successful completion of the course[,]?\s*",
        r"^after successful completion of this course[,]?\s*",
        r"^after successful passing the course[,]?\s*",
        r"^student will\s*",
        r"^students will be able to\s*",
]

    for pattern in additional_patterns:
        text = re.sub(pattern, "", text, flags=re.IGNORECASE)

    # oprava chybějících mezer po odrážkách
    text = re.sub(r"-(\w)", r"- \1", text)

    # odstranění strukturálního číslování položek seznamu
    text = re.sub(r"(^|\s)\d+\.\s+", " ", text)

    # opětovné sjednocení mezer po úpravách
    text = re.sub(r"\s+", " ", text).strip()

    # kontrola, zda po čištění nevznikla prázdná hodnota
    if text in ["", "-", ".", "_"]:
        return pd.NA

    return text


# aplikace čištění
df_vse[COLUMN] = df_vse[COLUMN].apply(clean_learning_outcomes)

df_partners[COLUMN] = df_partners[COLUMN].apply(clean_learning_outcomes)
```
